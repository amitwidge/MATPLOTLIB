```python
# Load library
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

#Create URL to CSV file (alternatively this can be a filepath)
wbst = 'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'  
    
# Load the first sheet of the JSON file into a data frame
df = pd.read_csv(wbst)
# View the first ten rows
df.head(10)
```

```python
# Pie-Chart

sums = df.sex.value_counts()
print(sums)
label_list = ["Male", "Female"]

plt.axis("equal")
plt.pie(
        sums,
        labels=label_list,
        autopct="%1.1f%%"
        )
plt.title("Male/Female Proportion")
plt.show()
```

```python
# Scatterplot

import matplotlib.colors as colors
import matplotlib.cm as cmx

x = df.fare
y = df.age

# Get unique names of sex
uniq = list(set(df['sex']))


# Set the color map to match the number of species
z = range(1,len(uniq))
hot = plt.get_cmap('hot')
cNorm  = colors.Normalize(vmin=0, vmax=len(uniq))
scalarMap = cmx.ScalarMappable(norm=cNorm, cmap=hot)



# Plot each species
for i in range(len(uniq)):
    indx = df['sex'] == uniq[i]
    plt.scatter(x[indx], y[indx], s=15, color=scalarMap.to_rgba(i), label=uniq[i])

plt.xlabel('Fare')
plt.ylabel('Age')
plt.title('Scatter Plot of Fare paid and Age')
plt.legend(loc='lower right')
plt.show()
```
